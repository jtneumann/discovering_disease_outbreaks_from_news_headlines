```python
headlinesFile = open('data/headlines.txt','r')
headlines = [line.strip()
             for line in headlinesFile.readlines()]
headlineCount = len(headlines)
print(f"{headlineCount} headlines on file.")
```

```python
import re
def name_via_regex(name):
    formatted_cityname = str(name)
    if name != formatted_cityname:
        regex = fr'\b({formatted_cityname})\b'
    else:
        regex = fr'\b{name}\b'
    return re.compile(regex, flags=re.IGNORECASE)
```

```python
from geonamescache import GeonamesCache
gc = GeonamesCache()

countries = [country['name'] for country in gc.get_countries().values()]
country_name = {name_via_regex(name): name for name in countries}

cities = [city['name'] for city in gc.get_cities().values()]
city_name = {name_via_regex(name): name for name in cities}
```

```python
def get_name_from_headlines(text, dictionary):
    for regex, name in sorted(dictionary.items(), key=lambda x: x[1]):
        if regex.search(text):
            return name
    return None
```

```python
import pandas as pd

matched_countries = [get_name_from_headlines(headline, country_name) for headline in headlines]
matched_cities = [get_name_from_headlines(headline, city_name) for headline in headlines]

data = {'Headline': headlines, 'City': matched_cities, 'Country': matched_countries}

df = pd.DataFrame(data)
```
